```python
import pandas as pd
import ast

# Load your dataset
df = pd.read_csv("your_file.csv")  # replace with your actual file name

# Function to extract only the first hero name
def extract_first_cast_name(cast_str):
    try:
        cast_list = ast.literal_eval(cast_str)
        if len(cast_list) > 0:
            return cast_list[0]['name']  # first hero name
        else:
            return None
    except:
        return None

# Apply the function
df['hero_name'] = df['cast'].apply(extract_first_cast_name)

# Keep only relevant columns
clean_df = df[['movie_id', 'title', 'hero_name']]

# Save to new CSV
clean_df.to_csv("cleaned_first_hero_name.csv", index=False)

print(clean_df.head())
```
