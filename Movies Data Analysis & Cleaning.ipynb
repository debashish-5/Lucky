```python
import pandas as pd
import ast

# Load your dataset
df = pd.read_csv("tmdb_5000_movies.csv")  # replace with your actual file name

# Function to safely parse JSON-like columns
def parse_json_column(value, key='name'):
    try:
        if pd.isna(value):
            return None
        data = ast.literal_eval(value)
        if isinstance(data, list):
            return ', '.join([d.get(key, '') for d in data if key in d])
    except:
        return None
    return None

# Clean specific columns
json_columns = ['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages']

for col in json_columns:
    df[col] = df[col].apply(parse_json_column)

# Convert release_date to proper datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Fill missing values
df.fillna({
    'homepage': 'Not Available',
    'tagline': 'No Tagline',
    'overview': 'No Overview Available'
}, inplace=True)

# Keep only important readable columns
clean_df = df[['id', 'title', 'original_title', 'genres', 'keywords',
               'production_companies', 'production_countries', 'spoken_languages',
               'release_date', 'runtime', 'budget', 'revenue', 'vote_average', 
               'vote_count', 'status', 'tagline', 'overview', 'popularity']]

# Save cleaned dataset
clean_df.to_csv("cleaned_movie_metadata.csv", index=False)

print("✅ Dataset cleaned successfully! Saved as 'cleaned_movie_metadata.csv'")
print(clean_df.head(10))
```
